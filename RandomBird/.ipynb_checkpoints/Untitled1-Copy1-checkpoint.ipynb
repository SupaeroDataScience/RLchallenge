```python
# You're not allowed to change this file
from ple.games.flappybird import FlappyBird
from ple import PLE
import numpy as np


def FlappyPolicy(statelist, screen, count_map):
    action=None
    if (statelist not in count_map[:,0].tolist()):
        return action
    index_state=count_map[:,0].tolist().index(statelist)
    if count_map[index_state, 2]>count_map[index_state, 1]:
        action=119
    return action

#list([256.,   0., 283., 148., 248., 427., 140., 240.]) in count_map[:,0].tolist()
#count_map[:,0].tolist().index(list([256.,   0., 283., 148., 248., 427., 140., 240.]))
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```

```python
#def update_reward_4_steps(count_map, memory):
 #   for i in memory:
  #      index_state=count_map[:,0].tolist().index(i[0])
   #     if i[1]==None:
    #        count_map[index_state,1]=-1
     #   else :
      #      count_map[index_state,2]=-1
            
#supprimer les etats par lequels on ne repassera pas : mettre une reward negative sur letat davant et supprimer letat dans lequel on sest fait niquer       
# prendre en compte le cas ou meme en sautant tu obtiens un truc qui ne fonctionne pas

# redondance des infos que le sol est mauvais peu importe la hauteur des tuyaux

# simule des etats terminaux qui remontent : un etat terminal artificiel est tel que tu choisisses de sauter ou pas tes niker.
# on arrive forcement par le bas puisque laction de base est none. Si on se crache, on essaye laction de sauter a letape juste 
# avant. Si ça ne fonctionne pas, on doit sauter encore avant.

# linformation du deuxieme tuyau ne sert a rien. 
# il faut les regrouper par meme premier tuyau afin qu il ne refasse pas les memes erreurs pour le meme premier tuyau
# mais un deuxieme tuyau different

# si le mec sest crashe en sautant, cest forcement un tuyau. SOit il set crashe en dessous du trou, soit au dessus
def update_reward_1_step(count_map, memory):
    if (memory[1][0] not in count_map[:,0]):
        index_state=count_map[:,0].tolist().index(memory[1][0])
        if memory[1][1] == None:
            count_map[index_state,1]=-1
        else :
            
            index_state=count_map[:,0].tolist().index(memory[0][0])
            count_map=np.delete(count_map,index_state,0)
        index_state=count_map[:,0].tolist().index(memory[2][0])
        count_map=np.delete(count_map,index_state,0)
    
```

```python
# Ajouter des valeurs a notre tableau, count_map est fait de telle sorte que la premiere colonne contient un etat,
#la seconde la valeur de Q pour laction None, 
# la troisieme la valeur de Q pour laction 119. Prends les valeurs a update et renvoie le tableau update :

# ne pad retenir les couples etats actions inuties; ie ou tu fais rien
def build_memory(count_map, statelist, action, reward):
    if action is None:
        action=1
    else:
        action=2
   
    bool1=(statelist not in count_map[:,0])
    if bool1 and (np.shape(count_map)[0]<100000) :
        a=[statelist,0.,0.]
        a[action]=reward*1.
        a=np.reshape(a,newshape=(1,3))
        count_map=np.concatenate((count_map,np.array(a)))
    if not bool1:
        index_state=count_map[:,0].tolist().index(statelist)
        count_map[index_state,action]+=reward*1.
    return count_map
        
# test OK :
#print(build_memory(np.reshape(a=np.array([0,0,0]),newshape=(1,3)),{'player_y': 256, 'player_vel': 0, 'next_pipe_dist_to_player': 283, 'next_pipe_top_y': 148, 'next_pipe_bottom_y': 248, 'next_next_pipe_dist_to_player': 427.0, 'next_next_pipe_top_y': 140, 'next_next_pipe_bottom_y': 240},None,1))
```
