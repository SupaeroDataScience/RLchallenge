```python
# Ajouter des valeurs a notre tableau, count_map est fait de telle sorte que la premiere colonne contient un etat,
#la seconde la valeur de Q pour laction None, 
# la troisieme la valeur de Q pour laction 119. Prends les valeurs a update et renvoie le tableau update :
def build_memory(count_map, statelist, action, reward):
    if action is None:
        action=1
    else:
        action=2
   
    bool1=(statelist not in count_map[:,0])
    if bool1 and (np.shape(count_map)[0]<100000):
        a=[statelist,0.,0.]
        a[action]=reward*1.
        a=np.reshape(a,newshape=(1,3))
        count_map=np.concatenate((count_map,np.array(a)))
    if not bool1:
        index_state=count_map[:,0].tolist().index(statelist)
        count_map[index_state,action]+=reward*1.
    return count_map
        
# test OK :
#print(build_memory(np.reshape(a=np.array([0,0,0]),newshape=(1,3)),{'player_y': 256, 'player_vel': 0, 'next_pipe_dist_to_player': 283, 'next_pipe_top_y': 148, 'next_pipe_bottom_y': 248, 'next_next_pipe_dist_to_player': 427.0, 'next_next_pipe_top_y': 140, 'next_next_pipe_bottom_y': 240},None,1))
```
